##‚Ä¢	How large is the gap in bank account usage between adult (18 and older) men and women? 

```python
# 2016: check for "Currently have"
df_2016['has_bank'] = df_2016[['e1_30', 'e1_31', 'e1_32']].apply(
    lambda row: 1 if 'Currently have' in row.values else 0, axis=1)

df_2016['has_mobile'] = df_2016['e1_37'].apply(
    lambda x: 1 if x == 'Currently have' else 0)

# 2021: check for "Currently use"
df_2021['has_bank'] = df_2021[['C1_30', 'C1_31', 'C1_28']].apply(
    lambda row: 1 if 'Currently use' in row.values else 0, axis=1)

df_2021['has_mobile'] = df_2021['C1_9'].apply(
    lambda x: 1 if x == 'Currently use' else 0)
```

```python
def weighted_mean(df, group_col, value_col, weight_col):
    return df.groupby(group_col).apply(
        lambda g: np.average(g[value_col], weights=g[weight_col])
    )

# Gender gap in bank & mobile money usage
print("2016 Bank:\n", weighted_mean(df_2016, 'gender_of_respondent', 'has_bank', 'popwgt_raw'))
print("\n2021 Bank:\n", weighted_mean(df_2021, 'gender', 'has_bank', 'IndWeight'))
print("\n2016 Mobile:\n", weighted_mean(df_2016, 'gender_of_respondent', 'has_mobile', 'popwgt_raw'))
print("\n2021 Mobile:\n", weighted_mean(df_2021, 'gender', 'has_mobile', 'IndWeight'))
```

2016: 34.3% (men) vs 21.5% (women) ‚Üí Gap = 12.8%

2021: 25.1% (men) vs 15.4% (women) ‚Üí Gap = 9.7%

‚Ä¢	Is this gender gap persistent across ages? 
1. Define age groups
2. Function to calculate weighted usage by gender + age

```python
# Define age bins and labels
bins = [17, 25, 35, 45, 55, 65, 200]
labels = ['18‚Äì25', '26‚Äì35', '36‚Äì45', '46‚Äì55', '56‚Äì65', '66+']

# Apply to both datasets
df_2016['age_group'] = pd.cut(df_2016['age'], bins=bins, labels=labels)
df_2021['age_group'] = pd.cut(df_2021['A19'], bins=bins, labels=labels)
```

```python
def weighted_usage_by_age(df, gender_col, age_col, usage_col, weight_col):
    return df.groupby([age_col, gender_col]).apply(
        lambda g: np.average(g[usage_col], weights=g[weight_col])
    ).unstack()
```

```python
print("üìä 2016 Bank Usage by Age & Gender:\n")
bank_2016_by_age = weighted_usage_by_age(df_2016, 'gender_of_respondent', 'age_group', 'has_bank', 'popwgt_raw')
print(bank_2016_by_age)

print("\nüìä 2021 Bank Usage by Age & Gender:\n")
bank_2021_by_age = weighted_usage_by_age(df_2021, 'gender', 'age_group', 'has_bank', 'IndWeight')
print(bank_2021_by_age)
```

‚Ä¢	Is this gender gap larger in rural or uban areas? 

```python
def weighted_usage_by_location(df, gender_col, location_col, usage_col, weight_col):
    return df.groupby([location_col, gender_col]).apply(
        lambda g: np.average(g[usage_col], weights=g[weight_col])
    ).unstack()
```

```python
print("üèòÔ∏è 2016 Bank Usage by Location & Gender:\n")
bank_2016_by_loc = weighted_usage_by_location(df_2016, 'gender_of_respondent', 'cluster_type', 'has_bank', 'popwgt_raw')
print(bank_2016_by_loc)

print("\nüèôÔ∏è 2021 Bank Usage by Location & Gender:\n")
bank_2021_by_loc = weighted_usage_by_location(df_2021, 'gender', 'cluster_type', 'has_bank', 'IndWeight')
print(bank_2021_by_loc)
```

Surprisingly, the gender gap in bank account usage is actually wider in urban areas in both 2016 and 2021.
This suggests that although urban areas have higher overall access, women may still be falling behind men in absolute growth.

Meanwhile, rural areas show lower usage for everyone, but the gender gap is slightly smaller ‚Äî likely because of overall structural barriers affecting both genders.

‚Ä¢	How has the gap changed since 2016? What was the main reason for the change in the gender gap in bank account usage between 2016 and 2021? 

```python
def summarize_gender_bank_usage(df, gender_col, usage_col, weight_col, extra_group=None):
    if extra_group:
        return df.groupby([extra_group, gender_col]).apply(
            lambda g: np.average(g[usage_col], weights=g[weight_col])
        ).unstack()
    else:
        return df.groupby(gender_col).apply(
            lambda g: np.average(g[usage_col], weights=g[weight_col])
        )
```

```python
# Overall drop
print("üîÑ Change in overall bank usage:")
print("2016:")
print(summarize_gender_bank_usage(df_2016, 'gender_of_respondent', 'has_bank', 'popwgt_raw'))
print("\n2021:")
print(summarize_gender_bank_usage(df_2021, 'gender', 'has_bank', 'IndWeight'))

# By location
print("\nüèòÔ∏è Change by location:")
print("2016:")
print(summarize_gender_bank_usage(df_2016, 'gender_of_respondent', 'has_bank', 'popwgt_raw', 'cluster_type'))
print("\n2021:")
print(summarize_gender_bank_usage(df_2021, 'gender', 'has_bank', 'IndWeight', 'cluster_type'))

# By age group
print("\nüìä Change by age:")
print("2016:")
print(summarize_gender_bank_usage(df_2016, 'gender_of_respondent', 'has_bank', 'popwgt_raw', 'age_group'))
print("\n2021:")
print(summarize_gender_bank_usage(df_2021, 'gender', 'has_bank', 'IndWeight', 'age_group'))
```

Bank usage declined for both men and women from 2016 to 2021.
But:

Women‚Äôs usage dropped less in some groups (e.g. older age)

The biggest male drops were in urban and working-age groups

This made the gender gap shrink ‚Äî not because women gained, but because men lost ground faster

### üìä Chart 1: Bank Account Ownership by Gender (2016 vs 2021)
Shows the proportion of men and women with bank accounts in both years.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style
sns.set(style="whitegrid")

# ----- VISUAL 1: Overall Gender Usage -----

# Create a DataFrame
bank_overall = pd.DataFrame({
    '2016': [0.215, 0.343],
    '2021': [0.154, 0.251]
}, index=['Female', 'Male'])

# Plot
bank_overall.T.plot(kind='bar', figsize=(8, 5), color=['#FF69B4', '#87CEFA'])
plt.title("Bank Account Ownership by Gender", fontsize=14)
plt.ylabel("Proportion")
plt.xticks(rotation=0)
plt.legend(title="Gender")
plt.ylim(0, 0.6)
plt.tight_layout()
plt.savefig('../charts/bank_account_ownership_by_gender.png', dpi=300)
plt.show()
```

### üìä Chart 2: Gender Gap in Bank Account Usage by Age Group
Visualizes how the gender gap (Male % - Female %) varies across age groups in 2016 and 2021.


```python
# Chart 2 - Gender Gap in Bank Usage by Age Group

# Age group labels
age_groups = ['18‚Äì25', '26‚Äì35', '36‚Äì45', '46‚Äì55', '56‚Äì65', '66+']

# Gender gaps by age (Male % - Female %)
gap_2016 = [0.266060 - 0.164157, 0.442799 - 0.296937, 0.400733 - 0.254525,
            0.406678 - 0.228158, 0.390254 - 0.183041, 0.248132 - 0.089135]

gap_2021 = [0.167978 - 0.114932, 0.322282 - 0.208072, 0.311196 - 0.179263,
            0.307114 - 0.167222, 0.296792 - 0.143871, 0.306102 - 0.179456]

# Plot
plt.figure(figsize=(8, 5))
plt.plot(age_groups, gap_2016, label='2016', marker='o', linestyle='--', color='#FF1493')
plt.plot(age_groups, gap_2021, label='2021', marker='o', linestyle='-', color='#1E90FF')

plt.title("Gender Gap in Bank Usage by Age Group", fontsize=14)
plt.ylabel("Gap (Male % - Female %)")
plt.xlabel("Age Group")
plt.ylim(0, 0.25)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Save before showing
plt.savefig('../charts/gender_gap_by_age.png', dpi=300)
plt.show()
```

### üìä Chart 3: Gender Gap in Financial Account Ownership (2016 vs 2021)
Compares the gender gap in bank account ownership and mobile money usage across 2016 and 2021. Gap is calculated as (Male % ‚àí Female %).


```python
import matplotlib.pyplot as plt
import numpy as np

# Chart 4 ‚Äì Gender Gap: Bank vs Mobile Money

years = ['2016', '2021']
bank_gap = [0.343 - 0.215, 0.251 - 0.154]      # Bank account gender gaps
mobile_gap = [0.720 - 0.653, 0.758 - 0.714]    # Mobile money gender gaps

x = np.arange(len(years))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, bank_gap, width, label='Bank Gap', color='#1E90FF')
bars2 = ax.bar(x + width/2, mobile_gap, width, label='Mobile Money Gap', color='#FF69B4')

# Labels
ax.set_ylabel('Gender Gap (Male ‚àí Female)', fontsize=12)
ax.set_title('Gender Gap in Financial Account Ownership', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.set_ylim(0, 0.15)
ax.legend()

# Add value labels
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.2%}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords='offset points', ha='center', fontsize=10)

plt.tight_layout()

# Save it!
plt.savefig('../charts/gender_gap_financial_accounts.png', dpi=300)
plt.show()
```

### üìä Chart 5: Trends in Bank vs Mobile Account Ownership by Gender (2016‚Äì2021)
